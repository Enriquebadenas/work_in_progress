```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

**Stocks in SP500**

```python
# Define the list of stock ticker symbols
ticker_symbols = ['QUBT', 'PPL', 'O']  # Example: Quantum Computing, Apple Inc., Microsoft

# Define the period you want to analyze
start_date = '2023-01-01'
end_date = '2025-01-17'

# Loop through each ticker symbol
for ticker_symbol in ticker_symbols:
    # Download the stock data
    stock_data = yf.download(ticker_symbol, start=start_date, end=end_date)

    # Calculate the 50-day and 200-day moving averages
    stock_data['SMA50'] = stock_data['Close'].rolling(window=50).mean()
    stock_data['SMA200'] = stock_data['Close'].rolling(window=200).mean()
    stock_data['Vol10'] = stock_data['Volume'].rolling(window=10).mean()

    # Plot the closing prices, moving averages, and trading volume for the stock
    fig, ax1 = plt.subplots(figsize=(12, 6))

    # Plot the closing prices and moving averages
    ax1.plot(stock_data['Close'], label='Closing Price')
    ax1.plot(stock_data['SMA50'], label='50-Day SMA', color='orange')
    ax1.plot(stock_data['SMA200'], label='200-Day SMA', color='red')
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Price')
    ax1.legend(loc='upper left')

    # Create a secondary y-axis to plot the trading volume
    ax2 = ax1.twinx()
    ax2.plot(stock_data.index, stock_data['Volume'], label='Volume', color='gray', alpha=0.3)
    ax2.plot(stock_data.index, stock_data['Vol10'], label='Vol.SMA10', color='green', alpha=0.4)
    ax2.set_ylabel('Volume')
    ax2.legend(loc='upper center')

    plt.title(f'{ticker_symbol} Stock Price, Moving Averages, and Volume')
    plt.grid()
    plt.show()
```
