Creación de gráficos Treemap (cajas de tamaño proporcionales al valor de una variable)

```python
!pip install squarify
import matplotlib.pyplot as plt
import squarify # pip install squarify (algorithm for treemap)
```

```python
# Change color
# en sizes =  [especificar el valor del área del rectángulo]
# en label = las etiquetas de texto
# alpha es la intensidad del color
# pad es la separación entre cajas
plt.figure(figsize=(10, 6))
plt.title('kW/día consumidos por Equipo. Tamaño proporcional al consumo')
squarify.plot(sizes=[13,22,35,5], label=["group A", "group B", "group C", "group D"], color=["red","green","blue", "grey"], alpha=.4, pad=0.2 )
plt.axis('off')
plt.savefig('Consumo diario por equipo recuadros.svg')
plt.show()
```
